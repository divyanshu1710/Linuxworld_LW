```python
import tkinter as tk
from tkinter import messagebox, filedialog
import webbrowser
import random

# Set up the GUI window
window = tk.Tk()
window.title("Interactive GUI")
window.geometry("400x400")

# Dark Mode toggle
is_dark_mode = False

def toggle_dark_mode():
    global is_dark_mode
    is_dark_mode = not is_dark_mode
    update_theme()

def update_theme():
    if is_dark_mode:
        window.config(bg="black", fg="white")
    else:
        window.config(bg="white", fg="black")

# Custom Themes
def apply_theme(theme):
    if theme == "blue":
        window.config(bg="blue", fg="white")
    elif theme == "green":
        window.config(bg="green", fg="white")
    elif theme == "pink":
        window.config(bg="pink", fg="black")

# Animated Transitions
def show_screen1():
    screen1_frame.pack()
    screen2_frame.pack_forget()

def show_screen2():
    screen1_frame.pack_forget()
    screen2_frame.pack()

# Interactive Buttons
def on_button_hover(event):
    event.widget.config(bg="red")

def on_button_leave(event):
    event.widget.config(bg="gray")

# Progress Indicator
def simulate_task():
    progress_bar.place(x=100, y=200)  # Place the progress bar on the window
    update_progress(0)  # Set the initial progress to 0
    window.after(1000, partial_progress)  # Start the partial progress simulation

def partial_progress():
    current_progress = progress_bar["value"]
    if current_progress < 100:
        update_progress(current_progress + random.randint(1, 10))
        window.after(1000, partial_progress)
    else:
        complete_task()

def complete_task():
    messagebox.showinfo("Task Complete", "Task has been completed.")
    progress_bar.place_forget()  # Remove the progress bar from the window

def update_progress(value):
    progress_bar["value"] = value
    progress_bar.update()

# Drag-and-Drop
def on_drag_start(event):
    event.widget.start_drag(event.x_root, event.y_root)

def on_drag_motion(event):
    event.widget.do_drag(event.x_root, event.y_root)

def on_drag_end(event):
    event.widget.end_drag()

# Keyboard Shortcuts
def handle_shortcut(event):
    if event.keysym == "n":
        messagebox.showinfo("Shortcut", "You pressed 'n' key!")

# Create the buttons and UI elements

# Dark Mode toggle button
dark_mode_button = tk.Button(window, text="Toggle Dark Mode", command=toggle_dark_mode)
dark_mode_button.pack()

# Custom Themes buttons
theme_blue_button = tk.Button(window, text="Blue Theme", command=lambda: apply_theme("blue"))
theme_blue_button.pack()

theme_green_button = tk.Button(window, text="Green Theme", command=lambda: apply_theme("green"))
theme_green_button.pack()

theme_pink_button = tk.Button(window, text="Pink Theme", command=lambda: apply_theme("pink"))
theme_pink_button.pack()

# Animated Transitions buttons
screen1_button = tk.Button(window, text="Screen 1", command=show_screen1)
screen1_button.pack()

screen2_button = tk.Button(window, text="Screen 2", command=show_screen2)
screen2_button.pack()

screen1_frame = tk.Frame(window, bg="white")
screen1_frame.pack()

screen2_frame = tk.Frame(window, bg="white")
screen2_frame.pack_forget()

# Interactive Buttons
interactive_button = tk.Button(window, text="Hover Me!")
interactive_button.pack()
interactive_button.bind("<Enter>", on_button_hover)
interactive_button.bind("<Leave>", on_button_leave)

# Progress Indicator
progress_bar = tk.Canvas(window, width=200, height=20, bg="lightgray", highlightthickness=0)
progress_bar.create_rectangle(0, 0, 0, 20, fill="blue", width=0)
progress_bar.place_forget()

simulate_task_button = tk.Button(window, text="Simulate Task", command=simulate_task)
simulate_task_button.pack()

# Drag-and-Drop
drag_label = tk.Label(window, text="Drag Me!", relief="solid")
drag_label.pack()
drag_label.bind("<ButtonPress-1>", on_drag_start)
drag_label.bind("<B1-Motion>", on_drag_motion)
drag_label.bind("<ButtonRelease-1>", on_drag_end)

# Keyboard Shortcuts
window.bind("<Key>", handle_shortcut)

# Run the Tkinter event loop
window.mainloop()
```
